## Object Oriented Programming in Python

#### OOPs (Object Oriented Programming)
OOPs (Object-Oriented Programming) is a programming approach based on objects and classes. It helps in organizing code into reusable and structured components.  
It is widely used to build scalable and maintainable applications.

---

**What is OOPs:**  :  a programming paradigm based on objects and classes  
  - Objects represent real-world entities with data (attributes) and behavior (methods)  
  - Classes are blueprints to create objects  
  - Helps in reusability, modularity, and better code structure  

---

**OOPs vs Procedural (functional) Programming:**  :  difference in approach  
  - Procedural programming focuses on functions and step-by-step execution  
  - OOPs focuses on objects and data with related behavior  
  - OOPs is better for large and complex applications  
  - Procedural is simpler and suitable for small tasks  

---

**How Python is Object Oriented Programming**

Python is an object-oriented programming (OOP) language, meaning it organizes code using objects and classes. 
Everything in Python are classes and objects.

#### Class, Object & Instances in Python

**What is a Class:**  A class is a blueprint used to create objects.  
  - It defines attributes (variables) and methods (functions).  
  - It does not hold actual data until objects are created.  
  - It is used to group related data and behavior.  

---

**What is an Object:**   An object is an instance of a class.  
- It represents real data based on the class.  
- It can access class attributes and methods.  
- It can have different values for each instance.  

---

**Defining a Class:**  :  A class is defined using the class keyword.  
**Syntax:**
```python
class ClassName:
    pass
```
---

**Creating Objects (Instances):**  :  An object is created by calling the class like a function.  
**Syntax:**
```python

```

---

**Class vs Instance:**  
A class is a blueprint, and an instance is a real object.  
A class defines structure, and an instance holds actual data.

#### Variables (Attributes) and it's type (Instance Variables & Class Variables)
Variables (Attributes) in a class are used to store data related to objects and the class.
There are two main types: instance variables and class variables.

---

**Instance Variables:**    They belong to each object separately.  
- Object owns the instance variable and  has its own copy of instance variables.  
- Instance variables are defined inside methods using self.
- They are usually initialized inside the constructor (__init__).  
**Syntax:**
```python
class ClassName:
    def __init__(self, value):
        self.value = value
```

**Class Variables:**  Class variables are defined inside the class but outside methods.  
- They are shared among all objects of the class.  
- Class owns the class variables (Object only can access them but they can't change them)
- All objects access the same value of class variables.  
- They are used for common data across instances.

**Syntax:**
```python
class ClassName:
    class_var = 10
```

---

- Key Points
  - instance variables are object-specific  
  - class variables are shared across all objects  
  - use instance variables for unique data and class variables for common data  

````python
# creating a class 
class Employee:
  # class variable (attributes)
  company_name = "NIIT"     # class owns it and only class can change it's value.

  def __init__(self, name, age):

    # instance variables (attributes)
    self.name = name    
    self.age = age    

  # instance method
  def print_employee_details(self):
    print(f"Name: {self.name}, Age: {self.age}, Company: {self.company_name}")



# creating object and passing values of attributes
e1 = Employee("Ayush", 23)
e2 = Employee("Gaurav", 24)

# accessing instance variable through object
print(e1.name, e1.age)

# accessing class variable though object
print(e1.company_name)

# accessing instance method through object
e1.print_employee_details()

# when we call the instance method, self got replaced by e1 (object)

# Note: we can access instance and class variables though object.

# accessing instance variable through class
# print(Employee.name)    # we can't instance variable through class. because instance variable is personal property of instance (object)

# accessing class variable though class
print(Employee.company_name)  # we can access class variable through class

# accessing instance method through class
Employee.print_employee_details(e1)
# we can access the instance method using class but we have to pass the object which will replace the self.


# modifying class variable using instance (object)
e1.company_name = "Coforge"
e1.print_employee_details()  # uses instance variable
e2.print_employee_details()  # still uses class variable
print(Employee.company_name) # unchanged

# Note: we can't change or modify the class variable using instance (object). But if we change the class variable using instance (object) then it will create a new instance variable for itself.


# we can change or modify the class variable using class
Employee.company_name = "Coforge Gr. Noida"

e2.print_employee_details()
e1.print_employee_details()   # now e1 also has it's own copy of company_name variable


# If class variable and instance variable has same name then instance variable will take precedence.
````

#### Methods (Instance, Class & Static Methods)
Methods are functions defined inside a class. They are used to perform operations using class or object data.  
There are three main types of methods based on how they access data.

---

**Instance Methods:**  :  Instance methods work with instance variables or object data.  
- They use self to access and change the instance variables.  
- They can also access class variables.  
- They are the most commonly used methods.  

**Syntax:**
```python
class ClassName:
    def method(self):
        pass
```
---

**Class Methods (@classmethod):**  :  Class methods work with class variables or class-level data.  
  - They use cls to access and change class variables.  
  - They cannot directly access instance variables.  
  - They are defined using @classmethod decorator.  

**Syntax:**
```python
class ClassName:
    @classmethod
    def method(cls):
        pass
```
---

**Static Methods (@staticmethod):**  :  Static methods do not access class or instance data.  
  - They behave like normal functions inside a class.  
  - They are used for utility logic related to the class.  
  - They are defined using @staticmethod decorator.  

**Syntax:**
```python
class ClassName:
    @staticmethod
    def method():
        pass
```
---

- Key Points
  - instance methods use self and work with object data  
  - class methods use cls and work with class data  
  - static methods do not depend on class or object  

````python
# creating a class
class Employee:
  company_name = "Coforge"

  # constructor is also the instance method
  def __init__(self, name, age):
    self.name = name
    self.age = age

  # instance method
  def print_employee_details(self):
    print(f"Name: {self.name}, Age: {self.age}, Company: {Employee.company_name}")

  # instance method to modify the instance variable
  def change_employee_name(self, new_name):
    self.name = new_name

  # class method
  @classmethod
  def change_company_name(cls, new_company_name):
    print(cls)
    cls.company_name = new_company_name

  # static method
  @staticmethod
  def print_greeting():
    print("Hello Happy Employee, Welcome to the company")


# creating object 
e1 = Employee("Ayush", 23)
e2 = Employee("Gaurav", 24)

# calling instance method to modify the instance variable
e1.change_employee_name("Aman")
e1.print_employee_details()

# calling class method to modify the class variable
Employee.change_company_name("Coforge Gr. Noida")
print(Employee.company_name)

# calling class method with object 
e1.print_employee_details()
e1.change_company_name("Coforge Gr. Noida, India")   # it will still change the class variable
e2.print_employee_details()

# calling static method though class
Employee.print_greeting()

# calling staticStat methid though object
e1.print_greeting()


# calling instance method through class
Employee.change_employee_name(e2, "Bhonu")
e2.print_employee_details()
````

| Via Object       | ✅ Yes         | ✅ Yes       | ✅ Yes        |
| Via Class        | ✅ Yes* (requires passing object manually)       | ✅ Yes       | ✅ Yes        |

#### Inheritance 
Inheritance allows one class to acquire properties and methods of another class. It helps in code reusability and reduces duplication. A class (child) inherits attributes and methods from another class (parent). It creates a hierarchy between classes.  
**Syntax:**
  ```python
  class Parent:
      pass

  class Child(Parent):
      pass
  ```

````python
# creating a parent class
class Parent:
  parent_variable = "I am variable of Parent"

  def parent_method(self):
      print("I am method of Parent")


# creating a child class which will inherit the parent class
class Child(Parent):
  child_variable = "I am variable of Child"

  def child_method(self):
      print("I am method of Child")


# creating object of child class
c1 = Child()

# accessing parent class variable through child class object
print(c1.parent_variable)

# accessing child class variable through child class object
print(c1.child_variable)

# accessing parent class method through child class object
c1.parent_method()

# accessing child class method through child class object
c1.child_method()

# but parent class objects can't access child class variable and method
````

#### Types of Inheritance in Python
Inheritance can be implemented in different forms based on how classes are related. It helps in structuring code and reusing functionality.  
Each type defines how child and parent classes are connected.

---

**Single Inheritance:**    A child class inherits from one parent class.
- It is the simplest form of inheritance.  
- It creates a one-to-one relationship.  

**Syntax:**
```python
class Parent:
    pass

class Child(Parent):
    pass
```
---

**Multiple Inheritance:**    A child class inherits from more than one parent class.  
- It allows combining features from multiple classes.  
- IIt creates a many-to-one relationship. 
- It follows method resolution order (MRO).  

**Syntax:**
```python
class Parent1:
    pass

class Parent2:
    pass

class Child(Parent1, Parent2):
    pass
```
---

**Multilevel Inheritance:**   A class inherits from a class which is already a child class.  
- It forms a chain of inheritance.  
- It allows step-by-step extension of features.  
- It creates a parent → child → grandchild structure.  

**Syntax:**
```python
class GrandParent:
    pass

class Parent(GrandParent):
    pass

class Child(Parent):
    pass
```
---

**Hierarchical Inheritance:**   Multiple child classes inherit from a single parent class.  
- It allows sharing common features among multiple classes.  
- It reduces code duplication.  
- It creates a one-to-many relationship.  

**Syntax:**
```python
class Parent:
    pass

class Child1(Parent):
    pass

class Child2(Parent):
    pass
```
---

**Hybrid Inheritance:** Combination of all the inheritances.

````python
# above example is the example of single inheritance

# let's see how does constructor works in inheritance

# case 1: When child class does not have constructor but parent class has constructor
# creating a parent class
class Parent:
  def __init__(self, name):
    print("Parent constructor called")
    self.parent_variable = name
  
  def parent_method(self):
      print("I am method of Parent")


# creating a child class which will inherit the parent class
class Child(Parent):
  def child_method(self):
      print("I am method of Child")

# creating object of child class. 
# When child class does not have constructor but parent class has constructor, Parent class constructor will be called automatically.
c1 = Child("Ayush")

# accessing parent class variable through child class object
print("Parent Variable:",c1.parent_variable)

c1.parent_method()
c1.child_method()
````

````python
# Case 2: When Parent and child class both have constructor
# creating a parent class
class Parent:
  def __init__(self, name):
    self.parent_variable = name
    print("Parent constructor called")

  def parent_method(self):
      print("I am method of Parent")


# creating a child class which will inherit the parent class
class Child(Parent):
  def __init__(self, name):
    print("Child constructor called")
    self.child_variable = name
  
  def child_method(self):
      print("I am method of Child")

# creating object of child class. 
# When Parent and child class both have constructor, Child class constructor will be called automatically.
# Child class constructor will override the Parent class constructor.
c1 = Child("Ayush")

# accessing parent class variable through child class object
# print("Parent Variable:", c1.parent_variable)  # we can't access it beause not initialized since parent class constructor is not called.

# accessing child class variable through child class object
print("Child Variable:", c1.child_variable)

c1.parent_method()
c1.child_method()
````

##### Constructor with super() method

If we write constructor in the both classes, parent class and child class then the parent class constructor is not available to the child class.    
In this case only child class constructor is accessible which means child class constructor is replacing parent class constructor.

**super()** method is used to call parent class constructor or methods from the child
class.

````python
class Person:
  def __init__(self, name, age):
    print("Parent class constructor called")
    self.name = name
    self.age = age

  def print_person_details(self):
    print(f"Name: {self.name}, Age: {self.age}")


class Employee(Person):
  def __init__(self, name, age, company, role):
    print("Child class constructor calles")
    super().__init__(name, age)   # calling parent class constructor
    self.company = company
    self.role = role

  def print_employee_details(self):
    print(f"Name: {self.name}, Age: {self.age}, Company: {self.company}, Role: {self.role}")


e1 = Employee("Ayush", 23, "Coforge", "Software Engineer")

e1.print_person_details()
e1.print_employee_details()
````

````python
# Multilevel Inheritance: A class inherits from a class which is already a child class.  
class Grandparent:
    def gp(self):
        print("Grandparent")

class Parent(Grandparent):
    def p(self):
        print("Parent")

class Child(Parent):
    def c(self):
        print("Child")

obj = Child()
obj.gp()
obj.p()
obj.c()
````

````python
# Hierarchical Inheritance: Multiple child classes inherit from a single parent class.  
class Parent:
    def show(self):
        print("Parent")

class Child1(Parent):
    pass

class Child2(Parent):
    pass

c1 = Child1()
c2 = Child2()

c1.show()
c2.show()
````

````python
# Multiple Ineritance:  A child class inherits from more than one parent class. 
class A:
    def show(self):
        print("A")

class B(A):
    def show(self):
        print("B")

class C(A):
    def show(self):
        print("C")

class D(B, C):
    pass


d = D()
d.show()

#  Method Resolution Order (MRO): It determines the sequence in which classes are searched. It starts from the current class and moves to parent classes. It follows a left-to-right order in inheritance. It uses C3 linearization algorithm for consistency. 

# check MRO
print(D.mro())
````

##### Method Overriding

When a child class defines a method with the same name as a method in the parent class, it overrides the parent class method.     
Note: If definition of parent and child method is different but name are same. child class method still overrides the parent method.

````python
# Method Overriding
class Parent:
  def show(self, a):
    print("Parent class method")

class Child(Parent):
  def show(self):
    print("Child class method")

c = Child()
c.show()
````

#### Encapsulation
Encapsulation is used to restrict direct access to data and control how it is modified. It helps in protecting data and maintaining clean code structure.  

---

**What is Encapsulation:**  :  It is the process of wrapping data and methods into a single unit (class).  
It restricts direct access to some variables and it improves data security and code maintainability.  

---

**Public, Protected, Private Attributes:**  
  :  Public attributes are accessible from anywhere.  
  :  Protected attributes use a single underscore (_) and should not be accessed directly.  
  :  Private attributes use double underscore (__) and are not directly accessible.  

**Syntax:**
```python
class ClassName:
    def __init__(self):
        self.public = 1
        self._protected = 2
        self.__private = 3
```
---

**Getters and Setters:**   Methods used to access and update private data.  
  - Getter is used to read value.  
  - Setter is used to modify value.  
  - **Syntax:**
    ```python
    def get_value(self):
        return self.__value

    def set_value(self, val):
        self.__value = val
    ```
---

**@property Decorator:**  Used to define getter method like an attribute.  
  - It allows access without calling method explicitly.  
  - It makes code cleaner and more readable.  
  - **Syntax:**
    ```python
    class ClassName:
        def __init__(self):
            self.__value = 0

        @property
        def value(self):
            return self.__value
    ```

````python
# Encapsulation:
class Employee:
    def __init__(self):
        self.name = "Aman"        # public
        self._salary = 50000      # protected
        self.__bonus = 10000      # private

    def show(self):
        print(self.name)
        print(self._salary)
        print(self.__bonus)


e = Employee()

# accessing variables
print(e.name)        # public
print(e._salary)     # protected (can access, but not recommended)

# print(e.__bonus)   # private (will give error)

# accessing private using method
e.show()
````

#### Method Overloading
Method overloading means defining multiple methods with the same name but different parameters. In Python, true method overloading is not directly supported.  
If we try to do , The last defined method overrides previous ones.  
It is achieved using default arguments or variable-length arguments (*args).  
It allows handling different types or number of inputs in a single method.


````python
class Math:
    def sum(self, a):
        return a
    
    # this will override the previous method
    def sum(self, a, b):
        return a + b
    
    # this will override the previous method
    def sum(self, a, b, c):
        return a + b + c
    
    # that's why we cant achive method overloading but we can use *args
    def sum(self, *args):
        return sum(args)
    
m1 = Math()
print(m1.sum(1, 2, 3))
print(m1.sum(1, 2))
print(m1.sum(1))
````

#### Operator Overloading (Dunder / Magic Methods)
Operator overloading allows operators to behave differently for user-defined objects. In Python, it is implemented using special methods called dunder (double underscore) methods.  
These methods define how operators work with objects.  
It helps customize operations like +, -, *, etc., for class objects.

- enables use of operators with custom objects  
- implemented using special methods like __add__, __sub__, etc.  
- improves readability and makes objects behave like built-in types  

**Syntax:**
```python
class ClassName:
    def __add__(self, other):
        # define addition logic
        return result
```

  `__add__`, `__sub__`, `__len__`, `__eq__`, `__ge__`, `__le__`, `__mul__`, etc are dunder magic methods in python. which can use for method overloading 

````python
# Example of operator overloading
class Complex:
  def __init__(self, real, imaginary):
    self.real = real
    self.imaginary = imaginary
  
  def __add__(self, other):
    real_result = self.real + other.real
    imaginary_result = self.imaginary + other.imaginary
    return Complex(real_result, imaginary_result)
  
  def __sub__(self, other):
    real_result = self.real - other.real
    imaginary_result = self.imaginary - other.imaginary
    return Complex(real_result, imaginary_result)
  
  def __str__(self):
    return f"Real: {self.real}, Imaginary: {self.imaginary}"
  
  def __eq__(self, other):
    if self.real == other.real and self.imaginary == other.imaginary:
      return True
    else:
      return False
  

c1 = Complex(1, 4)
c2 = Complex(-3, -7)
c3 = Complex(-3, -7)

print(c1)
c4 = c1.__add__(c2)  # c4 = c1 + c2
print(c4)
print(c1 - c2)
print(c3 == c2)
````

##### Difference between Method Overloading and Method Overriding

| Feature | Method Overloading | Method Overriding |
|--------|------------------|------------------|
| Definition | Same method name with different parameters | Same method name in parent and child class |
| Support in Python | Not directly supported | Fully supported |
| Involves Inheritance | Not required | Required |
| Parameters | Must be different | Must be same |
| Purpose | Handle different input cases | Change or extend parent behavior |
| Execution | Decided at compile-time (conceptually) | Decided at runtime |
| Method Resolution | Last method overrides previous ones | Child method overrides parent method |

#### Abstraction
Abstraction means hiding implementation details and showing only essential features. It helps reduce complexity and improves code design.  
In Python, abstraction is implemented using abstract classes and abstract methods.

---

**Abstract Classes:**  Abstract Class A class derived from ABC class which belongs to abc module, is known as abstract class in Python. ABC defines that the class which is derived from it becomes an abstract class.
  - An abstract class is a class that cannot be instantiated directly. We can't create objects of Abstract class.  
  - It is used as a blueprint for other classes.  
  - It can contain both abstract and normal methods.  

---

**Abstract Methods:**  An abstract method is a method declared but not implemented. It must be implemented by the child class.     We can declare a method as abstract method by using @abstractmethod decorator.
**Syntax:**
```python
from abc import ABC, abstractmethod

class Parent(ABC):
    @abstractmethod
    def show(self):
        pass
```
---

**Concert Methods:** Concrete Method A Concrete method is a method whose action is defined in the abstract class itself.

---

**Interface:** An interface is a blueprint for classes that defines a set of methods which must be implemented, but does not provide their implementation.  
Python does not have a built-in interface keyword, but interfaces can be created using abstract classes **with only abstract methods**.

````python

from abc import ABC, abstractmethod


# Define an abstract class
class Shape(ABC):
    def __init__(self, name):  # constructor
        self.name = name

    @abstractmethod
    def area(self):  # abstract method
        pass

    def shapeInfo(self):  # concrete method
        print(f"I am a {self.name}")


# Concrete subclass of the abstract class
class Circle(Shape):
    def __init__(self, name, radius):
        super().__init__(name)
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius


# Concrete subclass of the abstract class
class Square(Shape):
    def __init__(self, name, side):
        super().__init__(name)
        self.side = side

    def area(self):
        return self.side * self.side


sq = Square("Square", 4)
sq.shapeInfo()
print("Area of square:", sq.area())

cr = Circle("Circle", 7)
cr.shapeInfo()
print("Area of circle:", cr.area())
````

````python

# Define an Interface
class Shape(ABC):

    @abstractmethod
    def area(self):  # only abstract method
        pass


class Square(Shape):
    def __init__(self, side):
        self.side = side

    def area(self):
        return self.side * self.side


sq = Square(4)
print("Area of square:", sq.area())


````
