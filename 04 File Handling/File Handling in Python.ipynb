#### Directory/File Permission Modes (chmod)
Permission modes define who can read, write, or execute a file/directory. They are set using numeric (octal) values like 777, 754.  
 
Permissions are divided into three groups: owner, group, and others. Each group has read (r), write (w), and execute (x) permissions.  

**How these modes actually work**

Each digit is a sum of permissions:
- 4 = read (r)
- 2 = write (w)
- 1 = execute (x)  (Permission to run a program or file)

So:

- 7 = 4 + 2 + 1 → rwx
- 6 = 4 + 2 → rw-
- 5 = 4 + 1 → r-x
- 4 = 4 → r--
- 0 = nothing → ---

**What the 3 digits mean**

A mode like `755` has 3 parts:

```
[ Owner ][ Group ][ Others ]
    7        5        5     
```

So:

- Owner → 7 → rwx
- Group → 5 → r-x
- Others → 5 → r-x


````python
# change mode 
os.chmod("./my_dir", mode=0o757)

os.chmod("./my_dir", mode=750)
````

````python
# remove directories 

# rmdir(path) is a used to remove an empty directory
os.rmdir("./faltu")

# os.removedirs() is a Python method used to remove directories recursively (dirs must be empty)
# remove child if child is empty then remove parent if parent is empty
os.removedirs("./year 2026/month july/day 27")
````

````python
# remove all the dirs
ls = os.listdir("./my_dir/")

for dir in ls:
    os.rmdir(f"./my_dir/{dir}")
else:
    os.rmdir("./my_dir/")
````

#### File Handling in Python
File handling is used to read, write, and manage data stored in files. It helps in persistent data storage outside the program.  

---

**open() Function:**  
The open() function is used to open a file in a specific mode. It returns a file object to perform operations like read or write.  

- requires file name and mode  
- creates file if not exists (in write/append mode)  

**Syntax:**
```python
f = open("file.txt", "r")
```

---

**File Modes (r, w, a, rb, wb, r+):**
File modes define how a file is opened and what operations can be performed. Each mode controls reading, writing, or both.

* r → read (default, file must exist)
* w → write (overwrites file)
* a → append (adds data at end)
* rb → read in binary mode
* wb → write in binary mode
* r+ → read and write

**Syntax:**

```python
f = open("file.txt", "w")
```

---

**Closing a File (close()):**
The close() method is used to close an opened file. It frees system resources and ensures data is saved properly.

* always close file after use
* prevents data loss or corruption

**Syntax:**

```python
f.close()
```


````python
# opening a file in read mode
f = open("./hello.txt", 'r')

print(f)

# closing the file
f.close()
````

````python
# opening a file in read mode
f = open("./hello.txt", 'r')

file_content = f.read()

print(file_content)

# since we opened the file in read mode so we can't perform write operation
# f.write("hello")

f.close()
````

**Context Manager:**  
A context manager is used to manage resources automatically like opening and closing files. It ensures proper setup and cleanup without manual handling. 

A context manager is an object that manages resources using setup and cleanup steps. It automatically handles resource release after use.  

- avoids manual resource management  
- ensures proper cleanup even if error occurs  

**Syntax:**
```python
with open("file.txt", "r") as f:
    data = f.read()
```

---

**with Keyword (File Handling):**
The with keyword is used to work with context managers. It automatically opens and closes the file after execution.

* no need to call close() manually
* safer and cleaner way to handle files

**Syntax:**

```python
with open("file.txt", "w") as f:
    f.write("Hello")
```


````python
# Now we will be only using with keyword 
with open("./hello.txt") as f:
    print(f)
    file_content = f.read()
    print(file_content)
    # we don't need to close the file

````

##### Read Operation

| Operations                              | Syntax |
|-------------------------------------------|--------|
| Read entire file                          | file.read() |
| Read specific number of characters        | file.read(size) |
| Read one line                            | file.readline() |
| Read all lines as a list                 | file.readlines() |
| Reset file pointer to beginning          | file.seek(0) |
| Get current file pointer position        | file.tell() |

````python
# Reading files
with open("./hello.txt") as f:
    # file_content = f.read()
    print("-------- complete file content --------")
    print(file_content) 

with open("./hello.txt") as f:
    print("-------- starting 5 characters --------")
    five_char = f.read(5)
    print(five_char)

with open("./hello.txt") as f:
    print("-------- read line by line --------")
    first_line = f.readline()
    second_line = f.readline()
    print(first_line)
    print(second_line)

with open("./hello.txt") as f:
    print("-------- read all lines --------")
    line_list = f.readlines()
    print(line_list)
````

````python
# Note: Whenever we read a file pointer moves from its place. 
# we can get the position of pointer using tell() and change it using seek()

with open("./hello.txt", 'r') as f:
    first_five = f.read(5)

    print("-------- first five --------")
    print(first_five)
    print(f"Pointer position: {f.tell()}")

    print("-------- next five --------")
    next_five = f.read(5)
    print(next_five)
    print(f"Pointer position: {f.tell()}")

    print("-------- rest content --------")
    rest_content = f.read()
    print(rest_content)
    print(f"Pointer position: {f.tell()}")

    # that's how read line always gives the next line

    # resetting the pointer
    f.seek(0)
    print(f"Pointer position: {f.tell()}")

    # pointing at the any position
    f.seek(14)
    print(f"Pointer position: {f.tell()}")

````

| Operations                              | Syntax |

````python
# create a new file 
with open("./demo.txt", 'w') as f:
    pass
````

````python
# writing operations 
with open("./demo.txt", 'w') as f:
    f.write("Hello this is a new file.")

# 'w' : write mode overwrites the already existing content in the file. 
with open("./demo.txt", 'w') as f:
    f.write("This is the new content.")

# that's why use append method if you want to write in a file without losing the data.
````

````python
with open("./demo2.txt", 'a') as f:
    f.write("Hello There !!!")

# now old content will be there in the file.
````

````python
# remove the existing content
with open("./demo.txt", 'w') as f:
    f.truncate()
````

````python
with open("./hello.txt", 'r+') as f:
    content = f.read()
    print("--- old content ---")
    print(content)

    f.seek(0)  # move pointer to beginning
    f.write(content.upper())
    f.truncate()  # remove old content

    f.seek(0)  # go back again to read updated content
    print("--- new content ---")
    new_content = f.read()
    print(new_content)
````

#### shutil Module:
The shutil module provides functions to perform file operations easily. It works on top of OS module for advanced file handling.  

- used for copy, move, and delete operations  
- useful for file system automation tasks  

**Syntax:**
```python
import shutil
```

**Some common operations:**
| Functionality                              | Syntax |
|-------------------------------------------|--------|
| Copy file (content only)                  | shutil.copyfile(src, dst) |
| Copy file (with permissions)              | shutil.copy(src, dst) |
| Copy file (with metadata)                 | shutil.copy2(src, dst) |
| Copy entire directory                     | shutil.copytree(src, dst) |
| Move file or directory                    | shutil.move(src, dst) |
| Delete directory recursively              | shutil.rmtree(path) |
| Change file owner/group                   | shutil.chown(path, user=None, group=None) |
| Create archive (zip, tar, etc.)           | shutil.make_archive(base_name, format, root_dir) |
| Extract archive                           | shutil.unpack_archive(filename, extract_dir) |
