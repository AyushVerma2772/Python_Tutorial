#### Exception Handling in Python
Exception handling is used to handle runtime issues and prevent program crashes. It helps maintain smooth execution even when errors occur.  

---

**What is an Error:**  
An error is a problem in the program that prevents it from running correctly. It usually occurs due to syntax or logical mistakes.  

- occurs due to incorrect syntax or logic  
- stops program execution  
- must be fixed before running  

---

**What is an Exception:**  
An exception is an error that occurs during program execution. It can be handled so that the program does not stop abruptly.  

- occurs at runtime  
- can be handled using try-except  
- allows program continuation  

**Syntax:**
```python
try:
    # code
except Exception as e:
    # handle exception
````

---

**Types of Built-in Exceptions:**
Python provides many built-in exceptions to handle different error scenarios. Each exception represents a specific type of runtime issue.

* **TypeError** → wrong data type used
* **ValueError** → correct type but invalid value
* **IndexError** → accessing invalid index
* **KeyError** → accessing missing dictionary key
* **ZeroDivisionError** → division by zero


#### try-except Block in Python
The basic try-except block is used to handle general exceptions. Code inside try is executed, and if an error occurs, except block runs.  

**Syntax:**
```python
try:
    # code
except:
    # handle error
````

---

**Catching Specific Exceptions:**
Specific exceptions are caught by mentioning the exception type. It allows handling different errors in different ways. It avoids catching unwanted exceptions.

**Syntax:**

```python
try:
    # code
except ValueError:
    # handle value error
```

---

**Catching Multiple Exceptions:**
Multiple exceptions can be handled using multiple except blocks or a tuple. It helps manage different error types together.

* can use separate except blocks
* can group exceptions using tuple

**Syntax:**

```python
try:
    # code
except (ValueError, TypeError):
    # handle multiple errors
```

`````python
print("Starting of the code....")

a = 1500
b = 0

# we have to keep only that code which can generate error. because if there is any exception then it will not execute the remaining code.

try:
    result = a / b
    print(result)
    print("Remaining code of try block...")
except ZeroDivisionError:
    print("Error: Division by zero")


print("Ending of the code....")


`````

`````python
# catching multiple exception

a = 10
b = 'q'

try:
    result = a / b
    print(result)

# multiple exception
except ValueError:
    print("Error: Invalid input")
except ZeroDivisionError as zero_excep:       # we can also use alias
    print(zero_excep)
except TypeError:
    print("Error: Invalid input")
`````

`````python
# shorthand to catch multiple exception in a single except block
a = 10
b = 'q'

try:
    result = a / b
    print(result)
except (NameError, ValueError, TypeError, ZeroDivisionError) as excep_oj:
    print(f"Message: {excep_oj}")
`````


**Catching All Exceptions (broad except):**
A broad except catches all exceptions without specifying type. It should be avoided unless necessary.

* can hide real errors
* use only when exact exception is unknown

**Syntax:**

```python
try:
    # code
except:
    # handle all exceptions
```

---

**Exception Object (accessing error message):**
Exception object is used to capture and display error details. It helps in debugging and logging.

* gives error message
* useful for debugging

**Syntax:**

```python
try:
    # code
except Exception as e:
    print(e)
```

`````python
# Handling all exceptions 
a = 10
b = 'q'

try:
    result = a / b
    print(result)
except:
    print("Error Occurred")


# using Exception class to catch all exceptions (This is the best practice)
try:
    result = a / b
    print(result)
except Exception as excep_oj:
    print(f"Message: {excep_oj}")
`````

**else Block (when it runs):**  
The else block runs only if no exception occurs in the try block. It is used for code that should execute when everything works fine.  

**Syntax:**
```python
try:
    # code
except Exception:
    # handle exception
else:
    # runs if no exception
```

---

**finally Block (always runs):**
The finally block always runs regardless of whether an exception occurs or not. It is used for cleanup operations.

**Syntax:**

```python
try:
    # code
except Exception:
    # handle exception
finally:
    # always runs
```

---

**Combining try-except-else-finally:**
All blocks can be used together to handle exceptions and control execution properly. It ensures proper handling and cleanup.

* try → main code
* except → error handling
* else → runs if no error
* finally → always runs

**Syntax:**

```python
try:
    # code
except Exception:
    # handle exception
else:
    # runs if no exception
finally:
    # always runs
```

`````python
# combination of try - except - else - finally 
a = 10
b = 'q'

try:
    result = a / b
    print(result)
except (ZeroDivisionError, TypeError) as err_msg:
    print(f"Message: {err_msg}")
else:
    print("Code has executed successfully")
finally:
    print("This is finally block, It will run every time.")

`````

#### Creating and Raising Exceptions in Python
Custom exceptions are created to define application-specific errors. Raising exceptions helps control program flow when an unexpected condition occurs.  

---

**Creating Custom Exception:**  
A custom exception is created by defining a class that inherits from Exception. It allows defining your own error types for specific use cases.  

**Syntax:**
```python
class MyError(Exception):
    pass
```

---

**Raising an Exception:**
An exception is raised using the raise keyword when a condition is not met. It stops normal execution and transfers control to exception handling block.

* used to trigger an error manually
* helps validate conditions

**Syntax:**

```python
if condition:
    raise Exception("Error message")
```

---

**Raising Custom Exception:**
A custom exception can also be raised using the raise keyword. It allows handling specific application-level errors.

* used with user-defined exception class
* makes error handling more specific

**Syntax:**

```python
if condition:
    raise MyError("Custom error")
```


`````python
# creating own Exception
class InvalidBalance(Exception):
    pass


total_balance = 1000
withdraw_amount = 1500

try:
    if withdraw_amount > total_balance:
        raise InvalidBalance("Insufficient Balance")
    else:
        total_balance -= withdraw_amount
        print(f"Withdrawal successful !!!. Account balance: {total_balance}")
except InvalidBalance as err_msg:
    print(f"Message: {err_msg}")

`````

`````python
# rasing exception
a = 9
b = 0

try:
    if a != b:
        raise Exception("Error raised !!! Both Numbers are different.")
    else:
        print("both are equal")
except Exception as err_msg:
    print(f"Message: {err_msg}")
`````

#### Difference between Error, Exception and Warning

| Feature | Error | Exception | Warning |
|--------|-------|----------|---------|
| Definition | A serious issue that prevents program execution | A runtime issue that can be handled | A message indicating a potential problem |
| Occurrence | Usually at compile-time or due to syntax | Occurs during program execution | Occurs during execution as a caution |
| Handling | Cannot be handled | Can be handled using try-except | Can be controlled using warnings module |
| Impact | Stops program execution | May or may not stop execution | Does not stop execution |
| Purpose | Indicates critical failure | Indicates manageable issue | Alerts developer about future risks |
| Example | SyntaxError, IndentationError | ValueError, TypeError, IndexError | DeprecationWarning, UserWarning |
